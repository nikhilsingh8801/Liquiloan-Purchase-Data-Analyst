# Project Summary -

**Objective** :-
- To identify the key factors influencing the purchase of chocolate candy bars and generate actionable insights to optimize marketing strategies and improve sales performance.

**Data** **Description:-**

- The dataset contains information on chocolate candy bar purchases made by 500 individuals over a period of 2 years in a specific area. The data was collected through the use of loyalty cards at checkout in a physical FMCG store. It includes details about each individual's purchasing behavior, such as the frequency and quantity of chocolate candy bar purchases, along with any other relevant attributes linked to their loyalty card usage.

This data provides insights into consumer buying patterns and preferences for chocolate candy bars.

# So, according to the data set we have 2 years of data. so we can convert Days into Months.

```python
def Months_details(df):
  if df['Day'] in range(1,32):
   return 'January 2023'
  elif df['Day'] in range(32,61):
   return 'February 2023'
  elif df['Day'] in range(61,93):
   return 'March 2023'
  elif df['Day'] in range(93,124):
   return 'April 2023'
  elif df['Day'] in range(124,156):
   return 'May 2023'
  elif df['Day'] in range(156,187):
   return 'June 2023'
  elif df['Day'] in range(187,219):
   return 'July 2023'
  elif df['Day'] in range(219,251):
   return 'August 2023'
  elif df['Day'] in range(251,282):
   return 'September 2023'
  elif df['Day'] in range(282,314):
   return 'October 2023'
  elif df['Day'] in range(314,345):
   return 'November 2023'
  elif df['Day'] in range(345,377):
   return 'December 2023'

df['2023_Sales']= df.apply(Months_details,axis=1)
```

# Age_Category vs Incidence group by :-

```python
def Category_wise_Age(df):
    age = df['Age']
    if 18 <= age <= 30:
        return 'Young'
    elif 31 <= age <= 40:
        return 'Middle_age'
    elif 41 <= age <= 55:
        return 'early_old'
    elif 56 <= age <= 80:
      return 'Old'

df['Age_Category'] = df.apply(Category_wise_Age, axis=1)
```

```python
def without_promotion_price(df):
  if df['Brand'] == 1 and df['Promotion_1'] == 0:
    return df['Quantity'] * df['Price_1']
  elif df['Brand'] == 2 and df['Promotion_2'] == 0:
    return df['Quantity'] * df['Price_2']
  elif df['Brand'] == 3 and df['Promotion_3'] == 0:
    return df['Quantity'] * df['Price_3']
  elif df['Brand'] == 4 and df['Promotion_4'] == 0:
    return df['Quantity'] * df['Price_4']
  elif df['Brand'] == 5 and df['Promotion_5'] == 0:
    return df['Quantity'] * df['Price_5']


df['Total_Amount_Without_Promotion'] = df.apply(without_promotion_price, axis=1)
```

```python
def with_promotion_price(df):
  if df['Brand'] == 0 and df['Promotion_1'] == 0:
    return 0
  elif df['Brand'] == 1 and df['Promotion_1'] == 1:
    return df['Quantity'] * df['Price_1']
  elif df['Brand'] == 2 and df['Promotion_2'] == 1:
    return df['Quantity'] * df['Price_2']
  elif df['Brand'] == 3 and df['Promotion_3'] == 1:
    return df['Quantity'] * df['Price_3']
  elif df['Brand'] == 4 and df['Promotion_4'] == 1:
    return df['Quantity'] * df['Price_4']
  elif df['Brand'] == 5 and df['Promotion_5'] == 1:
    return df['Quantity'] * df['Price_5']
  elif df['Brand'] == 5 and df['Promotion_5'] == 0:
    return 0


df['Total_Amount_With_Promotion'] = df.apply(with_promotion_price, axis=1)
```
